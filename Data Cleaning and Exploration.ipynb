```python
## Standardizing Reporting Dates, Years, and Months
## &&
## Casting The Passenger Count
```

```python
for index, row in df.iterrows():
    up_value = str(row['Activity Period'])[:4]+'/'+str(row['Activity Period'])[-2:]
    df.at[index, 'Activity Period Converted'] = up_value
    df.at[index, 'Year']  = str(row['Activity Period'])[:4]
    df.at[index, 'Month'] = str(row['Activity Period'])[-2:]
    ## &&
    passenger_count = int(row['Passenger Count'].replace(',',''))
    df.at[index,'Passenger Count'] = passenger_count
```

```python
df['Passenger Count'] = df['Passenger Count'].astype(int)
```

```python
def calculating_passengers_sum(df, after_before, airline_name, period):
    condition = (df['Operating Airline'] == airline_name) | (df['Published Airline'] == airline_name)
    
    if after_before == 'before':
        total = df[(condition) & (df['Activity Period'] <= period)]['Passenger Count'].sum()
    else:
        total = df[(condition) & (df['Activity Period'] >= period)]['Passenger Count'].sum()

    return total
```

```python
def average_of_passengers(df,begin_year,end_year,airline_name):
    
    df_assist = df[ (df['Year']>=begin_year) & (df['Year']<=end_year) & ((df['Operating Airline'] == airline_name) | (df['Published Airline'] == airline_name))].groupby(by='Year').agg({'Passenger Count':'sum'})
    
    return df_assist['Passenger Count'].mean()
```
